```python
import struct
```

```python
abc = [16, 262, -2]
a = bytearray(struct.pack('h', int(abc[0])))
print(a[0], a[1])
a = bytearray(struct.pack('h', int(abc[1])))
print(a[0], a[1])
a = bytearray(struct.pack('h', int(abc[2])))
print(a[0], a[1])

data1 = "%02x%02x#" % (a[0], a[1])

print(data1)
data2 = struct.unpack('h', a)[0]
print(data2)
```
